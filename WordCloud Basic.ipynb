## wordcloud 만들기

```python
import matplotlib.pyplot as plt
%matplotlib inline

from wordcloud import WordCloud, STOPWORDS

text="""
28일 한국거래소 자료를 보면, 10월 들어 코스피는 315.92포인트(13.48%) 하락했다. 
코스닥지수도 같은 기간 159.20포인트(19.36%) 떨어졌다. 코스피 시가총액은 209조8510억원, 
코스닥 시가총액은 51조5290억원 줄었다. 한 달 새 국내 주식시장에서 투자자들이 보유한 주식 가치가 
261조3800억원 쪼그라든 것이다. 
이 같은 급락세는 2008년 10월 금융위기(-23.13%)나 
1997년 10월 외환위기(-27.25%) 때만큼은 아니지만 월간 기준으로 보면 2008년 10월 이후 가장 많이 하락했다.
다음달 한국은행의 기준금리 인상 가능성이 높게 점쳐지는 등 증시를 둘러싼 국내 사정도 녹록지 않다. 
NH투자증권은 이번주 코스피 밴드를 1960~2150으로 기존 전망치보다 크게 낮췄다. 단기적으로 저가 매수세가 유입될 수는 있지만 상승폭은 제한될 것으로 전망된다. 김병연 NH투자증권 연구원은 “현 가격 수준에서는 언제라도 기술적 반등이 나타날 수 있지만 상승 동력 부재로 상승폭은 제한될 것”이라고 말했다.
"""

spwords = set(STOPWORDS)  #기본적으로 많이 쓰는 제외할 단어
spwords.add("일")  # 제외하고 싶은 단어 추가

```

```python
from matplotlib import font_manager

font_dirs = ['/Library/Fonts']
font_files = font_manager.findSystemFonts(fontpaths=font_dirs)
font_list = font_manager.createFontList(font_files)
font_manager.fontManager.ttflist.extend(font_list)
```

```python
wordcloud = WordCloud (max_font_size = 200, font_path ='/Library/Fonts/Microsoft/Batang.ttf',
                      stopwords = spwords, background_color = '#FFFFFF',
                       width = 1200,
                       height = 800).generate(text)
```

```python
plt.figure (figsize = (10,8))
plt. imshow(wordcloud)
plt.tight_layout (pad = 0)

plt.savefig('wordcloud_image1.png', bbox_inches='tight') # 파일로 저장
plt.show()
```

## a.txt 파일의 내용으로 워드클라우드 생성

```python
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud, STOPWORDS


text = open('a.txt', 'r', encoding = 'utf-8').read()

spwords = set(STOPWORDS)  # 기본적으로 많이 쓰는 단어는 제외
spwords.add('10월')  # 제외하고 싶은 단어 추가

# 워드 클라우드를 설정한다. 한글이 있기 때문에 폰트 넣어주기

wordcloud = WordCloud (max_font_size = 200, font_path = '/Library/Fonts/Microsoft/Gulim.ttf',
                        stopwords = spwords, background_color = '#FFFFFF',
                        width = 1200, height = 800).generate(text)

plt.figure (figsize = (10, 8))
plt.imshow (wordcloud)
plt.tight_layout (pad = 0)
plt.axis ('off')


plt.savefig ('wordcloud_image2.png', bbox_inches = 'tight')
plt.show()
```

## 이미지 (mask.png) 파일을 밑그림, 바탕 이미지 모양으로 사용

```python
from IPython.display import Image
Image ('./mask.png')
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS

text = open ('a.txt', 'r', encoding = 'utf-8').read()

spwords = set(STOPWORDS)
spwords.add ('10월')

import numpy as np
from PIL import Image   #이미지 관련 함수

alice_mask = np.array(Image.open ('mask.png'))


#워드 클라우드 설정. 한글이 있기 때문에 폰트 넣어주기
wordcloud = WordCloud (max_font_size = 200, font_path = '/Library/Fonts/Microsoft/Gulim.ttf',
                       stopwords = spwords, background_color = '#FFFFFF',
                       max_words = 2000, mask = alice_mask, 
                       width = 1200, height = 800).generate(text)

plt.figure(figsize = (10,8))
plt.imshow (wordcloud)
plt.tight_layout (pad=0)
plt.axis ('off')
```

## colormap 사용하기

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS

text = open ('a.txt', 'r', encoding = 'utf-8').read()

spwords = set(STOPWORDS)   #기본적으로 많이 쓰는 단어
spwords.add('10월')   #제외하고 싶은 단어 추가

wordcloud = WordCloud (max_font_size = 200, font_path = '/Library/Fonts/Microsoft/Gulim.ttf',
                      stopwords = spwords, background_color = '#FFFFFF',
                      max_words = 2000,
                      colormap = 'spring_r',
                      width = 1200, height = 800).generate(text)

plt.figure (figsize = (10, 8))
plt.imshow (wordcloud)
plt.tight_layout(pad = 0)
plt.axis ('off')

plt.savefig ('wordcloud_image4.png', bbox_inches = 'tight')
plt.show()


```

## 리스트에 담겨있는 단어들의 개수를 세고, 많이 사용되는 단어별로 워드 클라우드 생성

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
from collections import Counter
```

```python
nouns = ['명사', '나', '선문', 'AI', '선문', '선문', '선문', '선문', '학교', '월요일', '월요일', '월요일',
        '수요일', '수요일', '수요일', '수요일']
```

```python
nouns = [ n for n in nouns if len(n) > 1]                 #2글자 이상만
print (nouns)
```

```python
count = Counter(nouns)      # 단어별 빈도수를 가진 딕셔너리를 만들어줌
print(count)

tags = count.most_common(3)      # 가장 많이 나타나는 3개 출력
print(tags)
```

```python
spwords = set(STOPWORDS)
spwords.add('학교')
wordcloud = WordCloud (max_font_size = 200, font_path = '/Library/Fonts/Microsoft/Gulim.ttf',
                      stopwords = spwords, background_color = '#FFFFFF',
                      max_words = 2000,
                      colormap = 'spring_r',
                      width = 1200, height = 800).generate_from_frequencies(dict(tags))

plt.figure(figsize = (10, 8))
plt.imshow (wordcloud)
plt.tight_layout (pad = 0)
plt.axis ('off')
plt.savefig ('wordcloud_image5.png', bbox_inches = 'tight')
plt.show()
```
